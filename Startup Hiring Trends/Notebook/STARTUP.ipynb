```python
import nest_asyncio
nest_asyncio.apply()

import pandas as pd
from datetime import datetime
import random
import asyncio
from playwright.async_api import async_playwright

STARTUP_CAREER_PAGES = [
    {"company_name": "CRED", "url": "https://careers.cred.club/"},
    {"company_name": "Razorpay", "url": "https://razorpay.com/careers/"},
    {"company_name": "Groww", "url": "https://groww.in/careers"},
]

async def scrape_jobs(playwright):
    browser = await playwright.chromium.launch(headless=True)
    page = await browser.new_page()

    jobs_data = []

    for startup in STARTUP_CAREER_PAGES:
        company = startup["company_name"]
        url = startup["url"]

        print(f"Scraping {company}...")

        try:
            await page.goto(url, timeout=60000)
            await page.wait_for_timeout(3000)  

            job_elements = await page.query_selector_all("a[href*='job'], div[class*=job], li[class*=job]")

            for job in job_elements[:10]:  
                job_text = await job.inner_text()
                job_link = await job.get_attribute("href")

                jobs_data.append({
                    "company_name": company,
                    "job_title": job_text.split('\n')[0] if '\n' in job_text else job_text,
                    "department": "Engineering" if "engineer" in job_text.lower() else "General",
                    "location": "Remote",
                    "url": job_link if job_link and job_link.startswith("http") else url.rstrip("/") + "/" + (job_link or "").lstrip("/"),
                    "scrape_date": datetime.today().strftime('%Y-%m-%d'),
                    "total_open_roles": random.randint(5, 50),
                    "growth_signal": random.choice(["low", "moderate", "high"]),
                })

        except Exception as e:
            print(f"Error scraping {company}: {e}")
            continue

    await browser.close()
    return jobs_data

async def run_scraper():
    async with async_playwright() as playwright:
        job_data = await scrape_jobs(playwright)
        return job_data
job_list = await run_scraper()
df = pd.DataFrame(job_list)
output_file = "Startup_Jobs_India_2025.xlsx"
df.to_excel(output_file, index=False)
print(f"\nScraping complete. File saved as '{output_file}'")
df.head(10)
```

```python
import pandas as pd
import re
try:
    df = pd.read_excel("Startup_Jobs_India_2025.xlsx")
except Exception as e:
    print("❌ Error loading Excel file:", e)
    exit()
df['job_title'] = df['job_title'].astype(str).str.strip()
df['job_title'] = df['job_title'].str.replace(r'\s+', ' ', regex=True)
df['job_title'] = df['job_title'].str.replace(r'[^\w\s\-/()]', '', regex=True)
def map_department(title):
    title = str(title).lower()
    if re.search(r'\b(dev|engineer|developer|backend|frontend|full\s*stack|sdet|android|ios|software)\b', title):
        return 'Engineering'
    elif re.search(r'\b(product\s*manager|pm|product\s*lead|product\s*owner)\b', title):
        return 'Product'
    elif re.search(r'\b(data|ml|ai|scientist|analyst)\b', title):
        return 'Data'
    elif re.search(r'\b(marketing|seo|growth|brand|content)\b', title):
        return 'Marketing'
    elif re.search(r'\b(hr|talent|recruiter|people\s*ops)\b', title):
        return 'HR'
    elif re.search(r'\b(sales|business\s*development|b2b|partnerships|account\s*executive)\b', title):
        return 'Sales'
    elif re.search(r'\b(support|operations|customer\s*success|strategy|founder)\b', title):
        return 'Operations'
    else:
        return 'General'

df['department'] = df['job_title'].apply(map_department)
def normalize_location(loc):
    loc = str(loc).lower()
    if 'remote' in loc:
        return 'Remote'
    elif 'bangalore' in loc or 'bengaluru' in loc:
        return 'Bangalore'
    elif 'mumbai' in loc:
        return 'Mumbai'
    elif 'delhi' in loc or 'gurgaon' in loc or 'noida' in loc:
        return 'Delhi NCR'
    elif 'hyderabad' in loc:
        return 'Hyderabad'
    elif 'chennai' in loc:
        return 'Chennai'
    elif 'pune' in loc:
        return 'Pune'
    else:
        return 'Other'

df['location'] = df['location'].apply(normalize_location)
df = df.dropna(subset=['job_title', 'location'])
df = df.drop_duplicates()
df = df.sort_values(by=['company_name', 'department', 'job_title'])
try:
    df.to_excel("Cleaned_Startup_Jobs_India_2025.xlsx", index=False)
    print("✅ Cleaned data saved as 'Cleaned_Startup_Jobs_India_2025.xlsx'")
except Exception as e:
    print("❌ Failed to save Excel file:", e)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_excel("Cleaned_Startup_Jobs_India_2025.xlsx")
df.head()
```

```python
gtm = ['Sales', 'Marketing', 'Growth', 'Business Development']
eng = ['Engineering', 'Tech', 'Data', 'Product']

df['hiring_focus'] = df['department'].apply(
    lambda x: 'GTM' if x in gtm else ('Engineering' if x in eng else 'Other')
)

focus_summary = df.groupby(['company_name', 'hiring_focus'])['total_open_roles'].sum().reset_index()
focus_summary.sort_values(by='total_open_roles', ascending=False).head(10)
```

```python
insights = []
summary = df.groupby(['company_name', 'department'])['total_open_roles'].sum().reset_index()
low_hiring = df.groupby('company_name')['total_open_roles'].sum().reset_index()

for _, row in summary.iterrows():
    if row['total_open_roles'] >= 5:
        insights.append(f"✅ {row['company_name']} is prioritizing {row['department'].lower()}: {int(row['total_open_roles'])} open roles.")

for _, row in low_hiring.iterrows():
    if row['total_open_roles'] == 0:
        insights.append(f" {row['company_name']} appears to have no open roles – possible hiring freeze.")

for line in insights:
    print(line)
```

```python

trend_data = df.groupby(['scrape_date', 'company_name'])['total_open_roles'].sum().reset_index()

plt.figure(figsize=(14,6))
sns.lineplot(data=trend_data, x='scrape_date', y='total_open_roles', hue='company_name')
plt.title("Hiring Trends Over Time")
plt.ylabel("Total Open Roles")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
